```python
#Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the dataset
df = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/CarPrice.csv")

df.head()
```

```python
#Cleaning the data
df.isnull().sum()
```

```python
df.describe()
```

```python
df.corr()
```

```python
encoded_data = pd.get_dummies(df, columns = ['CarName','enginetype', 'fueltype','cylindernumber','aspiration','doornumber','carbody','drivewheel','enginelocation','fuelsystem'])
print(encoded_data)
```

```python
# Data Visualization
sns.pairplot(df)
```

```python
plt.figure(figsize=(17, 17))
correlations = df.corr()
sns.heatmap(correlations, annot=True)
plt.show()
```

```python
# Segregating dataset into training and testing data
price = "price"
x = np.array(encoded_data.drop(["price",], 1))
y = np.array(encoded_data["price"])
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
# Training the model
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
```

```python
#Testing the Model
```

```python
y_pred = dtr.predict(x_test)
```

```python
y_pred
```

```python
# Testing the accuracy
from sklearn.metrics import mean_absolute_error
dtr.score(x_test, y_pred)*100
```
